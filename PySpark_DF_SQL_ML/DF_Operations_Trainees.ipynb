```python
from pyspark.sql.functions import *
import pyspark.sql.functions as F
```

```python
describe = describe.select('summary',F.round('Open', 2).cast('double').alias('Open'),
                            F.round('High', 2).cast('double').alias('High'),
                            F.round('Low', 2).cast('double').alias('Low'),
                            F.round('Close', 2).cast('double').alias('Close'),
                            F.round('Volume', 2).cast('double').alias('Volume'),
                            F.round('Adj Close', 2).cast('double').alias('Adj Close'))
```

```python
describe.show()
```

```python
# Notice all data types are Strings! 
df.describe().printSchema()
```

```python
from pyspark.sql.functions import format_number
```

```python
result = df.describe()
result.select(result['summary'],
              format_number(result['Open'].cast('float'),2).alias('Open'),
              format_number(result['High'].cast('float'),2).alias('High'),
              format_number(result['Low'].cast('float'),2).alias('Low'),
              format_number(result['Close'].cast('float'),2).alias('Close'),
              result['Volume'].cast('int').alias('Volume')
             ).show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
rate = df.select(df['High']/df['Volume']).withColumnRenamed("(High / Volume)", "HV Ratio")
```

```python
rate.show()
```

#### What day had the Peak High in Price?

```python
peak = df.select('Date','High').agg(F.max('High')).collect()[0][0]
```

```python
peak
```

```python
df.filter(df.High == peak).select("Date").collect()[0][0]
```

#### What is the mean of the Close column?

```python
df.select('Close').agg(F.avg('Close')).show()
```

#### What is the max and min of the Volume column?

```python
df.select('Volume').agg(F.max('Volume'),F.min('Volume')).show()
```

#### How many days was the Close lower than 60 dollars?

```python
df.filter(df['Close']<60).count()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
((df.filter(df['High']>80).count())/ df.count())*100
```

#### What is the Pearson correlation between High and Volume?

```python
df.select(corr("High", "Volume")).show()
```

#### What is the max High per year?

```python
df.groupBy(F.year('Date').alias('Year')).agg(F.max('High')).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months. 

```python
df.groupBy(F.month('Date').alias('Month')).agg(F.avg('Close')).orderBy('Month').show()
```
