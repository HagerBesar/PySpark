## Using SQL
### Create Temp View:

```python
df.createOrReplaceTempView("my_view")
```

### Display "RAM"column and count "RAM" column:

```python
df1 = spark.sql("select RAM, count(*) AS count from my_view GROUP BY RAM")

df1.show()
```

### Get all columns when "Year" column equal "2015"  

```python
df = spark.sql("select * from my_view where Year = 2015")

df.show()
```

### Get all when "Model" start with "M":

```python
result = spark.sql("SELECT Model FROM my_view WHERE Model LIKE 'M%'")

result.show()
```

### Get all data when "Model" column equal "MacBook Pro"

```python
df = spark.sql("SELECT * FROM my_view WHERE Model = 'MacBook Pro' ")

df.show()
```

### Get all data with Multiple Conditions when "RAM" column equal "8GB" and "Model" column is "Macbook".

```python
df1 = spark.sql("select * from my_view where Model = 'MacBook' and RAM='8GB' ")

df1.show()
```

### Get all data with Multiple Conditions when "D" greater than or equal "8" and "Model" column is "iMac".

```python
res = spark.sql("select * from my_view where D>=8 and Model = 'iMac' ")

res.show()
```
